```python
#e. ¿Cuántos años tiene el atleta más joven en nuestros datos de muestra?  Pista: usen min()
min_age = df['Age'].min()
print(f'El atleta mas joven tenia {min_age} de edad.')
```

```python
#a. ¿Cuántos atletas que participaron en el deporte "Tug-Of-War" (tira y afloja) pertenecían en el equipo “Sweden” (sueco) en el conjunto de datos de muestra?

tug_of_war = df['Sport'] == 'Tug-Of-War'
suecos = df['Team'] == 'Sweden'
suecosTUG = len(df[tug_of_war & suecos])
print(f"Hubo {suecosTUG}suecos en el Tug-Of-War.")
```

```python
# b. ¿Cuántas medallas de oro se le entregaron a los atletas que tenían más de 40 años?

filter = (df['Medal'] == 'Gold') & (df['Age'] > 40)
contadorcuarenta = len(df[filter])
print(f"Hubo {contadorcuarenta} medallistas de oro por encima de 40 años.")
```

```python
#3. Determinen la altura promedio de: Todas las filas incluidas-Todas las filas incluidas, pero agrupadas por Event (evento).

promedioedad = df['Height'].mean()
edadporevento = df.groupby('Event')['Height'].mean()
print(f'La altura promedio fue de {round(promedioedad, 1)}')
print()
print('Esta fue la altura promedio por evento')
print()
print(edadporevento)
```
